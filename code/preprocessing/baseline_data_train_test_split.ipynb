# Baseline model train test split 

## Setup & read in data 

```python
import numpy as np
import pandas as pd
from collections import Counter

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

from imblearn.over_sampling import SMOTE

from matplotlib import pyplot

import os
```

```python
os.getcwd()
```

```python
data_df = pd.read_csv('../../data/processed/combined_processed_data.csv')
```

```python
data_df.head()
```

## Data split 

```python
y = data_df['excluded']
X = data_df.drop(['Rndrng_NPI', 'excluded'], axis=1)
```

```python
## count of positive cases 
sum(y)
```

```python
# 70-15-15 train-validation-test split to have enough positive cases 

np.random.seed(123)
# Initial split 
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    stratify=y, 
                                                    test_size=0.15)
# Train vs validation split 
X_train, X_validation, y_train, y_validation = train_test_split(X_train, 
                                                                y_train, 
                                                                stratify=y_train,
                                                                test_size=0.176)
```

## SMOTE 

```python
oversample = SMOTE()
```

```python
X_train_smote, y_train_smote = oversample.fit_resample(X_train, y_train) 
X_validation_smote, y_validation_smote = oversample.fit_resample(X_validation, y_validation) 
X_test_smote, y_test_smote = oversample.fit_resample(X_test, y_test) 
```

## Scale values 

```python
stdsc1 = StandardScaler()

X_train = stdsc1.fit_transform(X_train) 
X_validation = stdsc1.transform(X_validation) 
X_test = stdsc1.transform(X_test) 
```

```python
stdsc2 = StandardScaler()

X_train_smote = stdsc2.fit_transform(X_train_smote) 
X_validation_smote = stdsc2.transform(X_validation_smote) 
X_test_smote = stdsc2.transform(X_test_smote) 
```
